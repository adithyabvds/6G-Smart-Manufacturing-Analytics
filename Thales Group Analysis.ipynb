### **Outlier Detection & Handling**

```python
# Numerical Features
numerical_columns = df.select_dtypes(include=['int64', 'int32', 'float64']).columns
numerical_columns
```

```python
# Boxplots
plt.figure(figsize=(20,18))
for i, column in enumerate(numerical_columns,1):
    plt.subplot(4,4,i)
    sns.boxplot(y=df[column], color='skyblue')
    plt.title(column)
plt.tight_layout()
plt.show()
```

```python
# IQR-Based Outlier Detection
outlier_summary = []
for column in numerical_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower) | (df[column] > upper)]
    outlier_summary.append({
        "Feature": column,
        "Outliers": len(outliers),
        "Percentage": round((len(outliers)/len(df))*100,2)
    })
outlier_df = pd.DataFrame(outlier_summary)
outlier_df
```

### **Machine Learning**

```python
# Feature Selection
X = df.drop(
    columns=[
        "Efficiency_Status",
        "Date",
        "Timestamp",
        "Weekday"
    ]
)

y = df["Efficiency_Status"]

print("Feature Shape :", X.shape)
print("Target Shape  :", y.shape)
```

```python
# Encode Categorical Variables
le_mode = LabelEncoder()
X["Operation_Mode"] = le_mode.fit_transform(X["Operation_Mode"])

le_target = LabelEncoder()
y = le_target.fit_transform(y)
```

```python
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
# Feature Scaling
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Build Logistic Regression Model
lr = LogisticRegression(
    max_iter=1000,
    random_state=42
)

lr.fit(X_train, y_train)
```

```python
# Make Predictions
lr_pred = lr.predict(X_test)
```

```python
# Prediction Probabilities
lr_prob = lr.predict_proba(X_test)

lr_prob[:5]
```

```python
# Model Accuracy
accuracy = accuracy_score(y_test, lr_pred)

print(f"Accuracy : {accuracy:.4f}")
```

```python
# Classification Report
print(classification_report(
    y_test,
    lr_pred,
    target_names=le_target.classes_
))
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, lr_pred)

plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=le_target.classes_,
    yticklabels=le_target.classes_
)

plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()
```

```python
# Compare Actual vs Predicted
prediction = pd.DataFrame({
    "Actual": le_target.inverse_transform(y_test),
    "Predicted": le_target.inverse_transform(lr_pred)
})

prediction.head(20)
```

```python
# Export Model
import joblib

artifacts = {
    "model": lr,
    "scaler": scaler,
    "operation_mode_encoder": le_mode,
    "target_encoder": le_target,
    "feature_names": X.columns.tolist()
}

joblib.dump(
    artifacts,
    "manufacturing_logistic_model.pkl"
)

print("Model exported successfully.")
```
